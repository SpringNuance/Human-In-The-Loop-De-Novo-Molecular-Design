### Finger prints and Tanimoto similarity

https://www.youtube.com/watch?v=3qzZbaUzo9M


```python
glycine = mol_list[0]
glycine
```

```python
bi = {}

fp = AllChem.GetMorganFingerprintAsBitVect(glycine, 2, nBits=1024, bitInfo=bi)
fp_arr = np.zeros((1,))
DataStructs.ConvertToNumpyArray(fp, fp_arr)
np.nonzero(fp_arr)
list(fp.GetOnBits())
```

```python
prints = [(glycine, x, bi) for x in fp.GetOnBits()]
Draw.DrawMorganBits(prints, molsPerRow=4, legends=[str(x) for x in fp.GetOnBits()])
```

```python
cysteine = mol_list[3]
img = Draw.MolsToGridImage([glycine, cysteine], molsPerRow = 4)
img
```

```python
fp2 = AllChem.GetMorganFingerprintAsBitVect(cysteine, 2, nBits=1024, bitInfo=bi)
print('cys', list(fp2.GetOnBits()))
print('gly', list(fp.GetOnBits()))

common = set(fp.GetOnBits()) & set(fp2.GetOnBits())
combined = set(fp.GetOnBits()) | set(fp2.GetOnBits())
print("common", common)
print("combined", combined)
print("common/combined", len(common)/len(combined))
print(DataStructs.TanimotoSimilarity(fp, fp2))
```

```python
data = pd.read_csv('delaney-processed.csv')[['Compound ID', 'smiles', 'measured log solubility in mols per litre']]
data.head()
```

```python
PandasTools.AddMoleculeColumnToFrame(data, 'smiles', 'Molecule')
data.head()
```

```python
data = data.drop(columns='smiles')
data.head()
```

```python
data.rename(columns={'measured log solubility in mols per litre': 'solub'}, inplace=True)
data.head()
```

```python
data.sort_values(by=['solub'], ascending=False).head()
```
